#### Integrating the given ODE to generate artificial data.

```python
# importing necessary libraries.
import numpy as np
import matplotlib.pyplot as plt
from scipy.integrate import solve_ivp
from scipy.interpolate import interp1d

# defining the function
def my_func(t,x):
    dx_dt = np.sin(x)
    return dx_dt

# inital conditons setting
x0 = [0.15]
tSpan= [0,10]
t = np.linspace(*tSpan,100)

# Integrating the ODE
intFunc = solve_ivp(my_func,tSpan, x0, dense_output = True)
x = intFunc.sol(t)
dx_dt = my_func(t,x)
x = np.squeeze(x)
dx_dt = np.squeeze(dx_dt)

```

```python
# data collect for gaussian
nData = 12
# not so good data
tDataNSG = np.linspace(*tSpan,nData)
xDataNSG = intFunc.sol(tDataNSG)
dx_dtDataNSG = my_func(t,xDataNSG)

xDataNSG = np.squeeze(xDataNSG)
dx_dtDataNSG = np.squeeze(dx_dtDataNSG)

# Better data collection
xMin = np.min(x)
xMax = np.max(x)
xRange = [xMin,xMax]

xData = np.linspace(*xRange,nData)
dx_dtData = my_func(t,xData)
xData = np.squeeze(xData)
dx_dtData = np.squeeze(dx_dtData)
#reshaping ofr GPR
xData = xData.reshape(-1,1)
dx_dtData = dx_dtData.reshape(-1,1)

# for the purpose of plotting only.
# what would tData be?
interpfun = interp1d(x,t)
tData=interpfun(xData)
x = x.reshape(-1,1)
```

```python
# Plotting.
# Why are we plotting?
# plotting the solution that is obtained from integration to verify if it looks reasonable to continue.
# Observe how data should be collected for GPR
fig,axs=plt.subplots(2,2,figsize=(10,10))

axs[(0,0)].set_title("X vs Time",fontweight="bold",fontsize=15)
axs[(0,1)].set_title("$\mathbf{\\frac{dx}{dt}}$ vs X",fontweight="bold",fontsize=15)
axs[(1,0)].set_title("X vs Time",fontweight="bold",fontsize=15)
axs[(1,1)].set_title("$\mathbf{\\frac{dx}{dt}}$ vs X",fontweight="bold",fontsize=15)

axs[(0,0)].set_xlabel("Time", fontweight="bold",fontsize=15)
axs[(0,1)].set_xlabel("X", fontweight="bold",fontsize=15)
axs[(1,0)].set_xlabel("Time", fontweight="bold",fontsize=15)
axs[(1,1)].set_xlabel("X", fontweight="bold",fontsize=15)

axs[(0,0)].set_ylabel("X",fontweight="bold",fontsize=15)
axs[(0,1)].set_ylabel("$\mathbf{\\frac{dx}{dt}}$",fontweight="bold",fontsize=15)
axs[(1,0)].set_ylabel("X",fontweight="bold",fontsize=15)
axs[(1,1)].set_ylabel("$\mathbf{\\frac{dx}{dt}}$",fontweight="bold",fontsize=15)

axs[(0,0)].plot(t,x,"r-",linewidth=2, label="Position evolution")
axs[(0,0)].plot(tDataNSG,xDataNSG, color = "brown", marker="o",linestyle = '',markeredgewidth=2,markerfacecolor="yellow", label="Uniform time data")
axs[(0,1)].plot(x,dx_dt,"b-",linewidth=2, label="Position evolution")
axs[(0,1)].plot(xDataNSG,dx_dtDataNSG,color = "brown", marker="o",linestyle = '',markeredgewidth=2,markerfacecolor="yellow", label="Uniform time data")

axs[(1,0)].plot(t,x,"r-",linewidth=2, label="Position evolution")
axs[(1,0)].plot(tData,xData,"go",markeredgewidth=2,markerfacecolor="yellow", label="Uniform distance data")
axs[(1,1)].plot(x,dx_dt,"b-",linewidth=2, label="Position evolution")
axs[(1,1)].plot(xData, dx_dtData,"go",markeredgewidth=2,markerfacecolor="yellow", label="Uniform distance data")

__ = [(0,0),(0,1),(1,0),(1,1)]
for _ in __:
    axs[_].set_facecolor("#f4f4f4")
    axs[_].grid(True,linestyle="--",alpha=0.6)
    spinesDict=["top","bottom", "left", "right"]
    for spineValue in spinesDict:
        axs[_].spines[spineValue].set_linewidth(2)
    axs[_].tick_params(length=5, width=2, direction="out",labelsize=12)
    
    legend=axs[_].legend(loc="best",fontsize=12)
    legend.get_frame().set_linewidth(2)
    legend.get_frame().set_facecolor("pink")
    legend.get_frame().set_edgecolor("black")

plt.subplots_adjust(wspace=0.35)
plt.subplots_adjust(hspace=0.35)
```

#### Care in data collection
Lets observe the above plots. Lets call the top left as (0,0), top right as (0,1), he bottom left as (1,0), the bottom right as (1,1).
- In plots (0,0), and (0,1)
    - The x vs time is used in collecting data.
    - The data has been collected uniformly in time and then corresponding velocities were collected.
    - We observe that such data collection leads to a lot of points collected at one place. There fore we are intrested in collecting data differently.
- In plots (1,0) and (1,1)
    - The $\frac{dx}{dt}$ vs time is used for data collection.
    - The correspondin points in (1,0) are just plotted for the sake of clarity and are not required at all.
 
#### Performing GPR on (1,1)
We see that (1,1) data collection strategy is better. We will use it for our GPR.

```python
# Importing the GPR libraries.
from sklearn.gaussian_process import GaussianProcessRegressor
from sklearn.gaussian_process.kernels import RBF

kernel = 1 * RBF(length_scale=1.0, length_scale_bounds=(1e-5, 1e5))
gaussian_process = GaussianProcessRegressor(kernel=kernel, n_restarts_optimizer=9)
gaussian_process.fit(xData, dx_dtData)
gaussian_process.kernel_
```

```python
# The mean function and the standard deviations
mean_prediction, std_prediction = gaussian_process.predict(x, return_std=True)
```

```python
# plotting GPR curves
fig,axs = plt.subplots(1,1,figsize=[8,8])

# Plotting the Phase trajectory
axs.set_title("GPR for $\mathbf{\\frac{dx}{dt}}$ vs X", fontweight = "bold", fontsize=15)
axs.set_xlabel("X", fontsize=15,fontweight="bold")
axs.set_ylabel("$\mathbf{\\frac{dx}{dt}}$",fontsize=20,fontweight="bold")
axs.set_facecolor("#f4f4f4")
axs.grid(True,linestyle="--", alpha=0.6)

spines_dir = ["top","bottom", "left", "right"]
for spine_value in spines_dir:
    axs.spines[spine_value].set_linewidth(2)
axs.tick_params(axis="both", length=5, width=2, direction="out",labelsize=15)

axs.plot(x, dx_dt, "r", linewidth = 4,label=r"Numerical Solution", linestyle="dotted")
axs.plot(x, mean_prediction,"b-", label="GPR Mean prediction")
axs.plot(xData, dx_dtData, "go", markeredgewidth =2, markerfacecolor = "yellow",label="Data to construct GPR")
axs.fill_between(
    x.ravel(),
    mean_prediction - 1.96 * std_prediction,
    mean_prediction + 1.96 * std_prediction,
    alpha=0.5,
    label=r"95% confidence interval",
)

legend = axs.legend(loc= "best",fontsize=15)
legend.get_frame().set_edgecolor("black")
legend.get_frame().set_linewidth(2)
legend.get_frame().set_facecolor("pink")
plt.savefig("4corGPRsolVeqSinx")
```

We can see that its matching surprisingly too nice. And the standard deviation is very little to get noticed in the graph. So assuming the mean function itself as the actual function, we will proceed further.  

Having known that $\frac{dx}{dt} = $ (mean function), we will use solve_ivp to find the new value of x.

# Integrating the mean function using solve_ivp

```python
# Since we are dealing with 1d case lets ignore the intricasies involved
# given an x it is giving us a velocity using GPR
def gprfun(t,x):
    x = x.reshape(-1,1)
    dx_dtGPR = gaussian_process.predict(x)
    return dx_dtGPR

gprfunIntegrate = solve_ivp(gprfun,tSpan,x0,dense_output=True)
xGPR = gprfunIntegrate.sol(t)
xGPR=np.squeeze(xGPR)
```

```python
# plotting the integrated curves
fig,axs = plt.subplots(1,1,figsize=[8,8])

# Plotting the Phase trajectory
axs.set_title("Position wrt GPR", fontweight = "bold", fontsize=15)
axs.set_xlabel("t", fontsize=15,fontweight="bold")
axs.set_ylabel("$X$",fontsize=20,fontweight="bold")
axs.set_facecolor("#f4f4f4")
axs.grid(True,linestyle="--", alpha=0.6)

spines_dir = ["top","bottom", "left", "right"]
for spine_value in spines_dir:
    axs.spines[spine_value].set_linewidth(2)
axs.tick_params(axis="both", length=5, width=2, direction="out",labelsize=15)

axs.plot(t, x, "r", linewidth = 4,label=r"Numerical Solution", linestyle="dotted")
axs.plot(t, xGPR,"b-", label="GPR Mean prediction")
axs.plot(tData, xData, "go", markeredgewidth =2, markerfacecolor = "yellow",label="Data to construct GPR")

legend = axs.legend(loc= "best",fontsize=15)
legend.get_frame().set_edgecolor("black")
legend.get_frame().set_linewidth(2)
legend.get_frame().set_facecolor("pink")
plt.savefig("4corVeqSinxPositionEstimate")
```

We observe that the integration performed is quite well and the mean function predicts the position very nicely.
