Let me define a 1 dimensional toy problem as follows:

$$\frac{dx}{dt}=sin(x)$$

Let us first see how its solution looks like (i.e., after integrating).

### Integrating the solution for the first t= 10 seconds with an initial condition of x= 0.25

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.integrate import solve_ivp

# Defining the function
def myfunc(t,x):
    dx_dt = np.sin(x)
    return dx_dt

#Setting up the initial conditions
time_span = [0,10]
X0 = [0.15]

# integrating the system
int_func = solve_ivp(myfunc,time_span,X0, dense_output='true')

# results 
t = np.linspace(*time_span, 100)
x = int_func.sol(t)
x = np.squeeze(x)
dx_dt = myfunc(t,x)
```

#### Collecting velocities at different time instances for GPR. 
Why we are doing this will be understood later.

```python
n_data = 12
t_data= np.linspace(*time_span,n_data)
x_data = int_func.sol(t_data)
v_data = myfunc(t_data,x_data)
v_data = np.squeeze(v_data)
```

```python
fig,axs = plt.subplots(1,3,figsize=[15,5])

# Plotting the Phase trajectory
axs[(0)].set_title("Phase Trajectory", fontweight = "bold", fontsize=15)
axs[(0)].set_xlabel("Time", fontsize=15,fontweight="bold")
axs[(0)].set_ylabel("x",fontsize=15,fontweight="bold")
axs[(0)].set_facecolor("#f4f4f4")
axs[(0)].grid(True,linestyle="--", alpha=0.6)

spines_dir = ["top","bottom", "left", "right"]
for spine_value in spines_dir:
    axs[(0)].spines[spine_value].set_linewidth(2)
axs[(0)].tick_params(axis="both", length=5, width=2, direction="out",labelsize=10)

axs[(0)].plot(t,x,"r-",linewidth=2,label="solution")

legend = axs[(0)].legend(loc= "best",fontsize=15)
legend.get_frame().set_edgecolor("black")
legend.get_frame().set_linewidth(2)
legend.get_frame().set_facecolor("pink")

# Plotting the velocity distribution
axs[(1)].set_title("Velocity vs X",fontweight="bold", fontsize=15)
axs[(1)].set_xlabel("X", fontweight="bold",fontsize=15)
axs[(1)].set_ylabel("$\mathbf{\\frac{dx}{dt}} $",fontweight = "bold", fontsize=15)
axs[(1)].set_facecolor("#f4f4f4")
axs[(1)].grid(True, linestyle="--", alpha=0.6)
for spine_value in spines_dir:
    axs[(1)].spines[spine_value].set_linewidth(2)

axs[(1)].tick_params(length=5,width=2,direction="out", labelsize=10)
axs[(1)].plot(x, dx_dt,"b-", linewidth=2,label="velocity")

legend = axs[(1)].legend(loc="best",fontsize=12)
legend.get_frame().set_facecolor("pink")
legend.get_frame().set_linewidth(2)
legend.get_frame().set_edgecolor("black")

# Plotting the Velocity vs time
axs[(2)].set_title("Velocity vs Time", fontweight = "bold", fontsize=15)
axs[(2)].set_xlabel("Time", fontsize=15,fontweight="bold")
axs[(2)].set_ylabel("x",fontsize=15,fontweight="bold")
axs[(2)].set_facecolor("#f4f4f4")
axs[(2)].grid(True,linestyle="--", alpha=0.6)

spines_dir = ["top","bottom", "left", "right"]
for spine_value in spines_dir:
    axs[(2)].spines[spine_value].set_linewidth(2)
axs[(2)].tick_params(axis="both", length=5, width=2, direction="out",labelsize=10)
 
axs[(2)].plot(t, dx_dt,"g-", linewidth=2,label="velocity")
axs[(2)].plot(t_data,v_data,"ro",markeredgewidth=2, markerfacecolor="yellow",label="data collected")

legend = axs[(2)].legend(loc="best",fontsize=12)
legend.get_frame().set_facecolor("pink")
legend.get_frame().set_linewidth(2)
legend.get_frame().set_edgecolor("black")


plt.subplots_adjust(wspace=0.25) 
```

## GPR implementation.

### Motivation 
Given the $\frac{dx}{dt}$ at different time instances, we would like to predict how solution using GPR.

### Plan for GPR implementation:
- Lets collect n_data =(12) velocity points from the velocity vs time plot shown above.The red outlined circles are the collected data.
- We will aplly GPR to estimate the green curve above.(which will be something else)
- Using the estimated (velocity vs time) obtained from GPR lets plot the phase trajectory and compare the results to see how GPR performs.

```python
# Importing the GPR libraries.
from sklearn.gaussian_process import GaussianProcessRegressor
from sklearn.gaussian_process.kernels import RBF

kernel = 1 * RBF(length_scale=1.0, length_scale_bounds=(1e-5, 1e5))
gaussian_process = GaussianProcessRegressor(kernel=kernel, n_restarts_optimizer=9)
gaussian_process.fit(t_data, v_data)
gaussian_process.kernel_
```
