```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D

x = np.linspace(-5,5,100)
y = np.linspace(-5,5,100)

X,Y = np.meshgrid(x,y)

Z = X**2 + Y**2

```

```python
axs={}
fig = plt.figure(figsize=(16,8))
axs[0] = fig.add_subplot(1,1,1,projection="3d")

axs[0].set_title("3D surface plot demonstration", fontweight="bold", fontsize=16)
axs[0].set_xlabel("X-Axis", fontweight="bold",fontsize=12)
axs[0].set_ylabel("Y-Axis",fontweight="bold",fontsize=12)
axs[0].set_zlabel("Z-Axis",fontweight ="bold",fontsize=12)

surf=axs[0].plot_surface(X,Y,Z,cmap='viridis')

fig.colorbar(surf, shrink=0.5, aspect=15)


plt.show()
```
