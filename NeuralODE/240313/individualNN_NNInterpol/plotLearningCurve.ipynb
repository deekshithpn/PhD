```python
import numpy as np
import matplotlib.pyplot as plt
file_path = 'learningCurveKurt.txt'

# Load data into a NumPy array
learning_rate = np.loadtxt(file_path,skiprows=1)

fig, ax = plt.subplots(figsize=(10, 8))

#ax.plot(x_train[:, 1], x_train[:, 2], x_train[:, 3], "k*", label="$given$")
plt.plot(learning_rate[:,0], learning_rate[:,1],"b",label='Loss')
# Customize fonts and sizes
ax.set_title("Learning Curve", fontsize=18, fontweight='bold')
ax.set_xlabel("Iterations", fontsize=14, fontweight='bold')
ax.set_ylabel("Loss", fontsize=14, fontweight='bold')

# Increase tick label font size
ax.tick_params(axis='both', labelsize=12)
ax.legend()
ax.grid(True, linestyle='--', alpha=0.6)
# Customize the appearance of the legend
legend = ax.legend(loc='upper right', fontsize=15)
legend.get_frame().set_edgecolor('black')
ax.set_facecolor('#F4F4F4')

plt.savefig('learningCurveKurt.pdf', bbox_inches='tight')

# Show the plot
plt.show()
```
