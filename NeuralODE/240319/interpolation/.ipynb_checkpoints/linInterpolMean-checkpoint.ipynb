```python
import numpy as np
import matplotlib.pyplot as plt
import scipy.io

matrix= scipy.io.loadmat("/home/ponnana/PhD/Yinling_copy/data_new.mat")

m = matrix["data_new"][:,0]
v =  matrix["data_new"][:,1]
s = matrix["data_new"][:,2]
k = matrix["data_new"][:,3]

tExp = [0.01058, 0.03158, 0.05258, 0.07358, 0.09458, 0.1156, 0.1366, 0.1576, 0.1786, 0.1996, 0.2206,
          0.2300, 0.2756, 0.3316, 0.3876, 0.4436, 0.4496, 0.5556, 0.6116, 0.6676, 0.7206, 0.7796, 0.8356,
          0.8916, 0.9476, 1.004, 1.060, 1.105, 1.124, 1.153, 1.182, 1.211, 1.240, 1.269, 1.298, 1.327,
          1.356, 1.385, 1.414, 1.443, 1.472, 1.501, 1.530, 1.559, 1.588, 1.617, 1.646, 1.670]

tExp = np.array(tExp)


t1,t2 = 0.24,1.1
xExp = np.concatenate((matrix["data_new"], tExp.reshape(-1,1)), axis=1)
# xExp = xExp.astype(double)

maxValues = np.max(xExp,axis=0)
maxValues[-1]=1
xExpNormal = xExp/maxValues
truncIndex = np.where(xExpNormal[:,1]>=0.0999)
xExpTruncNormal = xExpNormal[truncIndex,:]
xExpTruncNormal = np.squeeze(xExpTruncNormal)

tExpTrunc = xExpTruncNormal[:,-1]
```

Till now we have data points that are normalized. We will now interpolate these data points using linear, quadratic, cubic, cubic hermite, akima and see how they look like.

## Linear Interpolation

```python
from scipy.interpolate import interp1d
# data Linear interpolation
dataLIFunc = interp1d(tExpTrunc, xExpTruncNormal, kind = "linear", axis=0)
tTrain = np.linspace(np.min(tExpTrunc), np.max(tExpTrunc), 1000)
dataLI = dataLIFunc(tTrain) 
```

### Plots for Linear Interpolation


```python
def common_plot_code(axs):
    #Legend and grid
    legend = axs.legend(loc = "best", fontsize= 12)
    legend.get_frame().set_edgecolor("black")
    legend.get_frame().set_linewidth(2)
    legend.get_frame().set_facecolor("pink")
    axs.grid(True, linestyle = "--", alpha=0.6)
    axs.set_facecolor("#f4f4f4")
    #tickparams
    axs.tick_params(axis = "both",length=5,width=2, direction= "out", labelsize=10)
    #bordes
    spines_dir =["top","bottom", "left", "right"]
    for spines_value in spines_dir:
        axs.spines[spines_value].set_linewidth(2)
    return 
    
fig, axs = plt.subplots(2,2,figsize=(10,10))
plt.subplots_adjust(hspace=0.25)
##Mean Plot
axs[(0,0)].plot(tTrain, dataLI[:,0], "r-", label="Lin Interpol mean")
axs[(0,0)].plot(tExpTrunc, xExpTruncNormal[:,0], "ro",markeredgewidth = 2, markerfacecolor="yellow", linewidth=2,label = "groundTruth Mean")
#axis Titles
axs[(0,0)].set_title("Linear Interpolation of Mean", fontweight="bold",fontsize=12)
axs[(0,0)].set_xlabel("Time", fontweight="bold",fontsize=12)
axs[(0,0)].set_ylabel("Mean", fontweight="bold", fontsize=12)
common_plot_code(axs[(0,0)])
    
## Varince Plot
axs[(0,1)].plot(tTrain,dataLI[:,1], "b-", linewidth=2, label= "Lin interpol Variance")
axs[(0,1)].plot(tExpTrunc,xExpTruncNormal[:,1],"bo", markerfacecolor="yellow",markeredgewidth=1.5,label= "ground truth Variance")#markerfacecolor="red", fillstyle="none"
#axis Titles
axs[(0,1)].set_title("Linear Interpolation of Varince", fontweight="bold",fontsize=12)
axs[(0,1)].set_xlabel("Time", fontweight="bold",fontsize=12)
axs[(0,1)].set_ylabel("Variance", fontweight="bold", fontsize=12)
common_plot_code(axs[(0,1)])

## Skewness Plot
axs[(1,0)].plot(tTrain,dataLI[:,2], "g-", linewidth=2, label= "Lin interpol Skewness")
axs[(1,0)].plot(tExpTrunc,xExpTruncNormal[:,2],"go", markerfacecolor="yellow",markeredgewidth=1.5,label= "ground truth Variance")#markerfacecolor="red", fillstyle="none"
#axis Titles
axs[(1,0)].set_title("Linear Interpolation of Skewness", fontweight="bold",fontsize=12)
axs[(1,0)].set_xlabel("Time", fontweight="bold",fontsize=12)
axs[(1,0)].set_ylabel("Skewness", fontweight="bold", fontsize=12)
common_plot_code(axs[(1,0)])

## Skewness Plot
axs[(1,1)].plot(tTrain,dataLI[:,3], "k-", linewidth=2, label= "Lin interpol Kurtosis")
axs[(1,1)].plot(tExpTrunc,xExpTruncNormal[:,3],"ko", markerfacecolor="yellow",markeredgewidth=1.5,label= "ground truth Kurtosis")#markerfacecolor="red", fillstyle="none"
#axis Titles
axs[(1,1)].set_title("Linear Interpolation of Kurtosis", fontweight="bold",fontsize=12)
axs[(1,1)].set_xlabel("Time", fontweight="bold",fontsize=12)
axs[(1,1)].set_ylabel("Kurtosis", fontweight="bold", fontsize=12)
common_plot_code(axs[(1,1)])

```
